```
#filmes_plus.ipynb: é um programa no qual é o possível obter relatórios,correlações,gráficos e análises estatística de uma base de dados de filmes em .csv 
#Autoras= Bianca Suriano e Nayara Menezes 
#Email= bianca.suriano@aluno.ufabc.edu.br / nayara.ayllon@aluno.ufabc.edu.br 
#entrada: arquivo .csv com dados sobre filmes(Título,Ano,Duração,Nota e Diretor)
#saida:relatórios,análises estatísticas,correlação e gráficos dos dados.
#Ambiente de desenvolvimento: Google Colab (notebook)´
#última edição:30/11/2020 


#importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import sys 

# Importação e leitura do arquivo .csv
# link: https://drive.google.com/file/d/1jxVxT4303gjAev-OSrcn93k1i5MMNI1J/view?usp=sharing

codigo = "1jxVxT4303gjAev-OSrcn93k1i5MMNI1J"
file = "https://drive.google.com/u/3/uc?id=" + codigo + "&export=download"
filmes= pd.read_csv(file, sep = ",")

# Função menu() : o usuário digita um número através do input(),o número selecionado corresponde a alguma opção predefinida,na qual pode haver ou não necessidade de nova interação.
#opcao= seleciona uma ação do menu()
def menu():
    opcao=int(input('''
    	  Escolha o número no menu
1 - Relatório com uma coluna
2 - Relatório com até duas colunas
3 - Relatório com até três colunas
4 - Relatório com até quatro colunas
5 - Relatório com cinco colunas 
6 - Tabela completa
7 - Relatório completo com médias e medianas 
8 - Gráfico do tipo boxplot
9 - Gráfico do tipo histograma
10 - Correlação de colunas
11 - Análise estatística
0 - Fechar 
Escolha:'''))
  # opção 1: usuário deve digitar uma coluna para que seja gerado print() da mesma.
  #.title() possibilita que o usuário digite as colunas em qualquer formatação.(foi utilizado em todos os input())
  #rcoluna1=coluna selecionada para relatório
    if opcao == 1:
        rcoluna1=input("Digite uma coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
        print("**RELATÓRIO**")
        print()
        print(filmes[[rcoluna1]])
        print()

  #opção 0: mostra ao usuário uma mensagem de despedida e deixa o sistema atraves da função sys.exit()
    elif opcao == 0:
        print("Obrigado por utilizar nosso programa ;)")
        print('''
        Equipe Menezes & Suriano
        Feito por Bianca Suriano e Nayara Menezes''')
        sys.exit()

  #opção 2: o usuário pode digitar duas colunas (iguais ou diferentes) para queseja gerado um print das mesmas.
  #rcoluna1= a priemira coluna selecionada para relatório
  #rcoluna2= a segunda coluna selecionada para o relatório 
    elif opcao == 2:
         rcoluna1=input("Digite uma coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna2=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         print("**RELATÓRIO**")
         print()
         print(filmes[[rcoluna1,rcoluna2]])
         print()

  #opção 3: o usuário pode digitar três colunas (iguais ou diferentes) para que seja gerado um print() das mesmas.
  #rcoluna1= a priemira coluna selecionada para relatório
  #rcoluna2= a segunda coluna selecionada para o relatório 
  #rcoluna3= a terceira coluna selecionada para o relatório
    elif opcao == 3:
         rcoluna1=input("Digite uma coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna2=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna3=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         print("**RELATÓRIO**")
         print()
         print(filmes[[rcoluna1,rcoluna2,rcoluna3]])
         print()

  #opção 4: o usuário pode digitar quatro colunas (iguais ou diferentes) para que seja gerado um print() das mesmas.
  #rcoluna1= a priemira coluna selecionada para relatório
  #rcoluna2= a segunda coluna selecionada para o relatório 
  #rcoluna3= a terceira coluna selecionada para o relatório
  #rcoluna4=a quarta coluna selecionada para o relatório
    elif opcao == 4:
         rcoluna1=input("Digite uma coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna2=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna3=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna4=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         print("**RELATÓRIO**")
         print()
         print(filmes[[rcoluna1,rcoluna2,rcoluna3,rcoluna4]])
         print()

  #opção 5: o usuário pode digitar cinco colunas(iguais ou diferentes) para que seja gerado um print() das mesmas.
  #rcoluna1= a priemira coluna selecionada para relatório
  #rcoluna2= a segunda coluna selecionada para o relatório 
  #rcoluna3= a terceira coluna selecionada para o relatório
  #rcoluna4= a quarta coluna selecionada para o relatório
  #rcoluna5 = a quinta coluna selecionada para o relatório
    elif opcao == 5:
         rcoluna1=input("Digite uma coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna2=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna3=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna4=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         rcoluna5=input("Digite outra coluna para o Relatório [Título, Ano, Duração, Nota e Diretor]: →").title()
         print("**RELATÓRIO**")
         print()
         print(filmes[[rcoluna1,rcoluna2,rcoluna3,rcoluna4,rcoluna5]])
         print()

#opção 6: o usuario obtem um print() com o as cinco colunas 
    elif opcao == 6 :
         print("***Relatório***")
         print()
         print(filmes.columns)
         print()
         print(filmes)

#opção 7: o usuário obtem as médias e medianas das três colunas com valores numéricos.
#medias= médias das colunas numérica
#mediana= medianas das colunas numéricas  
    elif opcao == 7 :
         medias=filmes.mean()
         mediana=filmes.median()
         print()
         print("***Médias***")
         print(medias)
         print()
         print("***Medianas***")
         print(mediana)
         print()
    
#opção 10: o usuário pode correlacionar duas colunas a sua ecolha e será gerado um print() com o valor.
#coluna1=primeira coluna numérica para correlação 
#coluna2= segunda coluna numérica para correlação
    elif opcao == 10:
          coluna1 = input("Digite uma coluna correlação [Ano, Duração e Nota]:→ ").title()
          coluna2= input("Digite a segunda coluna para correlação [Ano, Duração e Nota]:→ ").title()
          cor = filmes[coluna1].corr(filmes[coluna2])
          print("Correlação entre",coluna1, "e", coluna2, ": %.4f" % cor)
          if  (cor > 0) :
             print("Correlação positiva")
          else:
             print('Correlação negativa')

#opção 11: O usuário pode digitar uma das colunas numéricas (Ano, Duração e Nota) e obter a média,mediana,desvio padrão e amplitude.
#Media= media da coluna numérica selecionada
#Mediana= mediana da coluna numérica selecionada 
#moda= moda da coluna numérica selecionada
#despd= desvio padrão da coluna numérica selecionada
#maximo= valor máximo para calculo da amplitude
#mínimo= valor mínimo para calculo da amplitude 
    elif opcao == 11 :
         acoluna1= input("Digite a coluna para análise estatística [Ano, Duração e Nota]:→ ").title()
         Media=filmes[acoluna1].mean()
         Mediana=filmes[acoluna1].median()
         moda=filmes[acoluna1].mode()
         despd=filmes[acoluna1].std()
         maximo= filmes[acoluna1].max()
         minimo= filmes[acoluna1].min()
         print()
         print("Média:",Media)
         print("Mediana",Mediana)
         print("Moda:",moda)
         print("Desvio padrão: %.2f" % despd)
         print("Amplitude:",(maximo - minimo),  "  Valor máximo:",maximo, " Valor mínimo:",minimo)

  #opção 8: O usuário pode digitar no input() uma das colunas numéricas e obter um gráfico do tipo boxplot.
  #gbcoluna1= coluna selecionada para o gráfico
    elif opcao == 8 :
          gbcoluna1=input("Digite a coluna para o gráfico do tipo boxplot [Ano, Duração e Nota]:→  ").title()
          filmes[gbcoluna1].plot(kind='box')
          pl.show()

  #opção 9: O usuário pode digitar no input() uma das colunas e obter um gráfico do tipo histograma.
  #ghcoluna1= coluna selecionada para o gráfico
    elif opcao == 9 :
         ghcoluna1=input("Digite a coluna para o gráfico do tipo boxplot [Ano, Duração e Nota]:→  ").title()
         print(filmes[ghcoluna1])
         filmes[ghcoluna1].plot(kind='hist')
         pl.show()

  #quando um número que não esta entre as opções é digitado em "opcao" será gerada uma mensagem pra o usuário que sera encaminhado ao início do programa novamente.
    else:
        print(''' 
        Poxa, este número não está nas alternativas :(
        Tente algum disponível no nosso Menu :)   ''')
        menu()

# após o usuário realizar as instruções da opção escolhida,ele será encaminhado novamente para menu()
# ValueError= caso seja inserida algum outro simbolo/letra na escolha da opção   
while True:
    try:
        menu()
    except ValueError:
        print('''
        Poxa,não é um número :(
        Tente novamente com uma das opções do nosso menu  :D''')    
   
```
